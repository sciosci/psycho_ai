```python
from psycho_ai import two_afc
import json
from itertools import product
import pickle5 as pickle
embedding = pickle.load(open('pubmed_w2v.pickle', 'rb'))
import matplotlib.pyplot as plt
def plot_pse_race(pse_score):
    plt.figure(figsize = (6, 6), dpi=150)
    plt.plot(list(pse_score.values()), list(pse_score.keys()), linewidth = 2, c = 'b')
    plt.gca().tick_params(axis='y', which='major', labelsize=12)
    plt.gca().tick_params(axis='x', which='major', direction="in", labelsize=8)

    plt.xticks([0, .5, 1], ['100% \n                 African-American Attribute', '50% African-American Attribute \n 50% European Attribute', '100% \n European Attribute              ']);
    # plt.yticks([0, 0.5, 1], ['No', '', 'Yes']);
    plt.xlabel('PSE', fontsize=15)
    plt.ylabel('Occupation', fontsize=15)
    # plt.title('Is it a European plumber?', fontsize=14);
    plt.gca().spines['right'].set_visible(False)
    plt.gca().spines['top'].set_visible(False)
    # Only show ticks on the left and bottom spines
    plt.gca().yaxis.set_ticks_position('left')
    plt.gca().xaxis.set_ticks_position('bottom')
    plt.grid(alpha=0.3)
    plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=1.0)

target_occupations = ['accountant', 'supervisor', 'worker', 'clerk', 'instructor', 'inspector',
          'electrician','appraiser', 'administrator', 'receptionist', 'advisor', 'chemist',
          'planner','paralegal', 'veterinarian', 'psychologist',  'baker', 'teacher', 
          'lawyer','nutritionist', 'hairdresser','pathologist', 'surgeon', 'practitioner', 
          'carpenter']
target_Research = ['manuscript', 'revision', 'editor', 'acceptance', 'rejection', 'high-profile', 
          'impact factor', 'research', 'thesis', 'dissertation']
target_teaching = ['class', 'grade', 'evaluation', 'course', 'semester', 'contents', 'teaching', 
          'syllabus', 'curriculum', 'credit']
target_service = ['committee', 'editor', 'reviewer', 'peer-reviewing', 'chair', 'conference', 
          'workshop', 'satellite session', 'symposium', 'summer school', 'sub-committee', 'panel', 
          'reviewer', 'organizer', 'advisor']
target_funding = ['grant', 'funding', 'career', 'national science foundation', 'national institutes of health', 'funding agency', 'foundation', 
          'budget', 'overhead', 'federal']



female_male_pairs = [['woman', 'man'],
                     ['female', 'male'],
                     ['she', 'he'],
                     ['her', 'him'],
#                    ['hers', 'his'],
                     ['daughter', 'son'],
                     ['girl', 'boy'],
                     ['sister', 'brother']]
names = json.load(open('race.json','r'))
african_european_pair = list(product(names['african-american'],names['european']))
```

```python
pse_score = two_afc.pse(embedding, target_Research, female_male_pairs)
two_afc.plot_pse(pse_score)
```
